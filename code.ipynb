```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set custom theme for Seaborn plots
sns.set_theme(style="ticks")
sns.set_style("whitegrid")

print("All modules initialized")

# Load the heart failure dataset
heart_data = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(heart_data.head(5))

# Display the last 5 rows of the dataset
print("Last 5 rows of the dataset:")
print(heart_data.tail(5))

# Check for duplicate entries in the dataset
has_duplicates = heart_data.duplicated().any()
print(f"Are there any duplicates in the dataset? {has_duplicates}")

# Data cleaning and transformation
cleaned_data = pd.DataFrame(heart_data)
cleaned_data["age"] = heart_data["age"]
cleaned_data["anaemia"] = np.where(heart_data["anaemia"] < 1, "No", "Yes")
cleaned_data["diabetes"] = np.where(heart_data["diabetes"] < 1, "No", "Yes")
cleaned_data["ejection_fraction"] = heart_data["ejection_fraction"]
cleaned_data["high_blood_pressure"] = np.where(heart_data["high_blood_pressure"] < 1, "No", "Yes")
cleaned_data["platelets"] = heart_data["platelets"]
cleaned_data["serum_creatinine"] = heart_data["serum_creatinine"]
cleaned_data["serum_sodium"] = heart_data["serum_sodium"]
cleaned_data["sex"] = np.where(heart_data["sex"] < 1, "Female", "Male")
cleaned_data["smoking"] = np.where(heart_data["smoking"] < 1, "No", "Yes")
cleaned_data["death_event"] = np.where(heart_data["DEATH_EVENT"] < 1, "No", "Yes")

# Display the cleaned DataFrame
print("Cleaned DataFrame:")
print(cleaned_data.head())

# Check the structure of the cleaned DataFrame
cleaned_data.info()

# Remove the 'time' column as it is not needed
heart_data.drop(['time'], axis=1, inplace=True)
print("Successfully removed 'time' from the dataset.")

# Summary statistics of the dataset
print("Summary statistics:")
print(heart_data.describe())

# Display the shape of the cleaned DataFrame
print(f"Shape of the cleaned DataFrame: {cleaned_data.shape}")

# Display value counts for categorical variables
print("Value counts for sex:")
print(cleaned_data["sex"].value_counts())
print("Value counts for high blood pressure:")
print(cleaned_data["high_blood_pressure"].value_counts())
print("Value counts for diabetes:")
print(cleaned_data["diabetes"].value_counts())
print("Value counts for smoking:")
print(cleaned_data["smoking"].value_counts())
print("Value counts for death event:")
print(cleaned_data["death_event"].value_counts())

# Visualize relationships in the dataset
sns.set()
sns.pairplot(cleaned_data, hue="death_event", palette="Set2")
plt.show()

sns.set()
sns.pairplot(cleaned_data, hue="sex", palette="Set2")
plt.show()

# Bar chart for male/female frequencies
plt.figure(figsize=(5, 5))
cleaned_data["sex"].value_counts(ascending=True).plot.barh(color=["#FF6F61", "#6B5B95"])
plt.title("Frequencies of Male and Female")
plt.ylabel("Sex")
plt.xlabel("Number of Candidates")
plt.show()

# Bar chart for death event frequencies
cleaned_data["death_event"].value_counts(ascending=True).plot.barh(color=["#FF6F61", "#6B5B95"])
plt.title("Death Event Frequencies")
plt.ylabel("Death Event")
plt.xlabel("Number of Deaths")
plt.show()

# Bar chart for death event per sex
death_counts_per_sex = cleaned_data.groupby("sex")["death_event"].value_counts().unstack().fillna(0)
death_counts_per_sex.plot.bar(color=["#FF6F61", "#6B5B95"])
plt.title("Death Event per Sex")
plt.xlabel("Sex")
plt.ylabel("Number of Deaths")
plt.show()

# Heatmap for correlation between sex and death event
sns.heatmap(pd.crosstab(heart_data["sex"], heart_data["DEATH_EVENT"]), cmap='Pastel1', linewidth=.5, square=True, center=0)
plt.title("Correlation Between Sex and Death Event")
plt.xlabel("Death Event")
plt.ylabel("Sex")
plt.show()

# Relationship between age and platelets, differentiated by sex
sns.lmplot(x="age", y="platelets", hue="sex", data=cleaned_data).set(title='Relationship Between Age and Platelets by Sex')
plt.show()

# Bar chart for smoking against death event
cleaned_data.groupby("smoking")["death_event"].value_counts().unstack().plot.barh(color=["#FF6F61", "#6B5B95"])
plt.title("Impact of Smoking on Death Events")
plt.xlabel("Number of Deaths")
plt.ylabel("Smoking Status")
plt.show()

# Relationship between high blood pressure and age, differentiated by sex
sns.catplot(x="high_blood_pressure", y="age", hue="sex", data=cleaned_data).set(
    xlabel="High Blood Pressure Status",
    ylabel="Age of Candidates",
    title="Relationship Between High Blood Pressure and Age by Sex"
)
plt.show()

# Count plot for death events
sns.catplot(x="death_event", data=cleaned_data, kind="count", palette="Pastel1")
plt.title("Count of Death Events")
plt.xlabel("Death Event")
plt.ylabel("Count")
plt.show()
```
